# OmniFinan Unified API Demo

Notebook replacement for the old `unified_api.py` `__main__` sample code.

```python
from omnifinan.unified_api import (
    get_company_news,
    get_financial_metrics,
    get_insider_trades,
    get_market_cap,
    get_price_df,
    get_prices,
    search_line_items,
)
```

```python
cases = [
    ("US", "AAPL"),
    ("HK", "00700"),
    ("CN", "600519"),
]

for market, ticker in cases:
    print(f"\\n--- {market} {ticker} ---")
    try:
        prices = get_prices(ticker, start_date="2024-01-01", end_date="2024-04-01")
        print("prices:", len(prices))
        df = get_price_df(ticker, start_date="2024-01-01", end_date="2024-04-01")
        print(df.tail(2))
    except Exception as exc:
        print("prices error:", exc)

    try:
        metrics = get_financial_metrics(ticker)
        print("financial_metrics:", len(metrics))
    except Exception as exc:
        print("financial metrics error:", exc)

    try:
        cap = get_market_cap(ticker)
        print("market_cap:", cap)
    except Exception as exc:
        print("market cap error:", exc)

    try:
        news = get_company_news(ticker, limit=1)
        print("news:", len(news))
    except Exception as exc:
        print("news error:", exc)

    try:
        insider = get_insider_trades(ticker, limit=1, end_date="2024-04-01")
        print("insider_trades:", len(insider))
    except Exception as exc:
        print("insider trades error:", exc)

    try:
        items = search_line_items(ticker, limit=2)
        print("line_items:", len(items))
    except Exception as exc:
        print("line items error:", exc)
```
