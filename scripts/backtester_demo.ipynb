# OmniFinan Backtester Demo

Notebook replacement for the old `backtester.py` `__main__` interactive script.

```python
from omnifinan.backtester import Backtester
from omnifinan.core.workflow import run_hedge_fund
```

```python
tickers = ["AAPL", "MSFT"]
start_date = "2025-01-01"
end_date = "2025-03-31"
selected_analysts = [
    "technical_analyst",
    "fundamentals_analyst",
    "macro_analyst",
    "sentiment_analyst",
    "valuation_analyst",
]
```

```python
backtester = Backtester(
    agent=run_hedge_fund,
    tickers=tickers,
    start_date=start_date,
    end_date=end_date,
    initial_capital=100000,
    model_name="deepseek-chat",
    provider_api="deepseek",
    selected_analysts=selected_analysts,
    initial_margin_requirement=0.0,
    annual_risk_free_rate=0.0434,
)
```

```python
performance_metrics = backtester.run_backtest()
performance_metrics
```

```python
performance_df = backtester.analyze_performance()
performance_df.tail()
```
